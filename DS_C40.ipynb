```python
# Numeric to Numeric
```

```python
sns.scatterplot(x = "GrLivArea", y = 'SalePrice', data = dataframe)
```

```python
sns.lmplot(x = "GrLivArea", y = 'SalePrice', data = dataframe)
```

```python
sns.jointplot(x = "GrLivArea", y = 'SalePrice', data = dataframe, kind = "hex")
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(dataframe.corr(), cmap="YlGnBu")
```

```python
k = 10
```

```python
cols = dataframe.corr().nlargest(k,'SalePrice')["SalePrice"].index
cols
cm = dataframe[cols].corr()
sns.heatmap(cm, cmap="YlGnBu")
```

```python
cols
```

```python
sns.pairplot(dataframe[cols])
```

```python
# Numerical vs Categorical
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(y = "Neighborhood", x = "SalePrice", data = dataframe)
```

```python
plt.figure(figsize=(15, 8))
sns.swarmplot(x = "OverallQual", y = "SalePrice", data = dataframe, size = 2)
```

```python
# Categorical vs Categorical
```

```python
crosstab = pd.crosstab(index = dataframe["Neighborhood"], columns = dataframe["OverallQual"])
crosstab
```

```python
crosstab.plot(kind = "bar", stacked = True, figsize = (12, 8))
```
